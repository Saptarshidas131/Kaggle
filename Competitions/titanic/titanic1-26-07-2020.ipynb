### Loading Datasets

#### Loading Training Data

```python
df_train = pd.read_csv('/kaggle/input/titanic/train.csv')
df_train.head()
```

#### Loading Training Data

```python
df_test = pd.read_csv('/kaggle/input/titanic/test.csv')
df_test.head()
```

#### Percent of survival of women in train data

```python
women = df_train.loc[df_train.Sex == 'female']['Survived']
rate_women = sum(women)/len(women)
print("Survival percent of women is {}%".format(rate_women*100))
```

#### Percent of survival of men in train data

```python
men = df_train.loc[df_train.Sex == 'male']['Survived']
rate_men = sum(men)/len(men)
print("Survival percent of men is {}%".format(rate_men*100))
```

#### Training a random forest model with 100 trees

```python
# import RandomForestClassifier from scikitlearn
from sklearn.ensemble import RandomForestClassifier

# training data, has to predict
y = df_train["Survived"]

# features used to predict
features = ["Pclass",  "Sex", "SibSp","Parch"]
# convert categorical data into indicator variables
X = pd.get_dummies(df_train[features])
X_test = pd.get_dummies(df_test[features])

# making the random forest classifer model
rfmodel = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
#fitting the data
rfmodel.fit(X,y)
# making predictions on X_test
predict = rfmodel.predict(X_test)
```

#### Convert to csv

```python
# create a python datframe output from predict by using dictionary
output = pd.DataFrame({'PassengerId': df_test.PassengerId, 'Survived': predict})
# creating a submission_titanic.csv from the dataframe
output.to_csv('submission_titanic.csv', index=False)
```
